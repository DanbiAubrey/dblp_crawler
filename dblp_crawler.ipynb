## TKDE beautifulsoup

```python
from bs4 import BeautifulSoup as soup
import requests
```

```python
year = 2021
# conf_name = 'tkde'
# volume = "https://dblp.org/db/journals/tkde/tkde34.html"

# conf_name = 'cscw'
# volume = "https://dblp.org/db/conf/cscw/cscw{}c.html".format(year)#CSCW

# conf_name = 'chi'
# volume = "https://dblp.org/db/conf/chi/chi{}.html".format(year)#CHI

# conf_name = 'iswc'
# volume = "https://dblp.org/db/conf/iswc/iswc{}.html".format(year)#ISWC

# conf_name = 'ubicomp'
# volume = "https://dblp.org/db/conf/huc/ubicomp{}ap.html".format(year)#UbiComp

conf_name = 'wsdm'
volume = "https://dblp.org/db/conf/wsdm/wsdm{}.html".format(year)
```

```python
page = requests.get(volume)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
uls = soup.select('ul.publ-list')
with open('titles_only/{}{}.txt'.format(conf_name,year), 'w') as f:
    for ul in uls:
        titles = ul.select('li > cite > span.title')
        for title in titles:
            f.write(title.get_text().rstrip(".") + "\n")
            print(title.get_text())
```
