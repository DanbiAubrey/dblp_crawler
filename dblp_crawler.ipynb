## TKDE beautifulsoup

```python
from bs4 import BeautifulSoup as soup
import requests
```

```python
url_tkde = "https://dblp.org/db/journals/tkde/tkde34.html"
```

```python
page = requests.get(url_tkde)
```

```python
soup_tkde = BeautifulSoup(page.content, "html.parser")
```

```python
uls = soup_tkde.select('ul.publ-list')
with open('titles_tkde.txt', 'w') as f:
    for ul in uls:
        titles = ul.select('li > cite > span.title')
        for title in titles:
            f.write(title.get_text().rstrip(".") + "\n")
            print(title.get_text())
```
