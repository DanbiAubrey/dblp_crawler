## WordCloud

```python
year = '2021'
# conf_name = 'tkde'
# conf_name = 'cscw'
# conf_name = 'chi'
# conf_name = 'iswc'
# conf_name = 'ubicomp'
conf_name = 'wsdm'

strings = []
with open("titles_only/{}{}.txt".format(conf_name, year), "r") as file:
    strings = [x.strip().rstrip(".") for x in file.readlines()]
    print(strings)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
```

```python
stopwords = set(STOPWORDS)
comment_words = ""
```

```python
for word in strings:
    tokens = word.split()
    
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
        
    comment_words += " ".join(tokens)+" "
```

```python
print(comment_words)
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)

wordcloud.to_file('wordcloud_figs/{}{}.png'.format(conf_name, year))
# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
